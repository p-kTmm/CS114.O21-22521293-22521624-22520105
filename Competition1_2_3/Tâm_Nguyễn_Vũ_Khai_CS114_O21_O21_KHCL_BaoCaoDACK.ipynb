  * Code tạo đặc trưng Dataset 2 (df_main là annonimized.csv):
    ```[python]
          import pandas as pd
          import numpy as np

          # Lấy giờ từ chuỗi 'created_at' bằng cách tách chuỗi
          def extract_hour(created_at):
              try:
                  return int(created_at.split()[1].split(':')[0])
              except:
                  return np.nan

          df_main['hour'] = df_main['created_at'].apply(extract_hour)

          # Tổng hợp dữ liệu
          agg_df = df_main.groupby("concat('it001', username)").agg(
              unique_assignment_id=("concat('it001',`assignment_id`)", 'nunique'),
              unique_problem_id=("concat('it001',`problem_id`)", 'nunique'),
              num_submit=("concat('it001', username)", 'count'),
              avg_pre_score=('pre_score', 'mean')
          ).reset_index()

          # Tính toán tỷ lệ hoàn thành bài tập trên 80%
          max_unique_assignment_id = agg_df['unique_assignment_id'].max()
          max_unique_problem_id = agg_df['unique_problem_id'].max()

          agg_df['completion_rate1'] = agg_df['unique_assignment_id'] / max_unique_assignment_id
          agg_df['completion_rate2'] = agg_df['unique_problem_id'] / max_unique_problem_id

          # Thêm hai cột thể hiện giá trị 1 nếu tỷ lệ hoàn thành bài tập lớn hơn hoặc bằng 80%, ngược lại là 0
          agg_df['above_80_assignment'] = (agg_df['completion_rate1'] >= 0.8).astype(int)
          agg_df['above_80_problem'] = (agg_df['completion_rate2'] >= 0.8).astype(int)

          # Tính toán tỷ lệ số lần nộp trung bình của mỗi sinh viên
          agg_df['num_submit_per_unique_assignment_id'] = (agg_df['num_submit'] + 10**-9) / agg_df['unique_assignment_id']
          agg_df['num_submit_per_unique_problem_id'] = (agg_df['num_submit'] + 10**-9) / agg_df['unique_problem_id']

          # Tạo vector tần suất 24 phần tử
          hourly_activity = df_main.groupby(["concat('it001', username)", 'hour']).size().unstack(fill_value=0)
          hourly_activity.columns = [f'activity_hour_{i}' for i in range(24)]

          # Kết hợp với dataframe chính
          agg_df = agg_df.merge(hourly_activity, on="concat('it001', username)", how='left')

          # Mã hóa 'assignment_id' để lưu số lượng nộp bài cho mỗi assignment
          assignment_binary = df_main.pivot_table(index="concat('it001', username)", columns="concat('it001',`assignment_id`)", aggfunc='size', fill_value=0)
          assignment_binary.columns = [f'assignment_{col}' for col in assignment_binary.columns]

          # Kết hợp với dataframe chính
          agg_df = agg_df.merge(assignment_binary, on="concat('it001', username)", how='left')

          # Mã hóa 'problem_id' để lưu số lượng nộp bài cho mỗi problem
          problem_binary = df_main.pivot_table(index="concat('it001', username)", columns="concat('it001',`problem_id`)", aggfunc='size', fill_value=0)
          problem_binary.columns = [f'problem_{col}' for col in problem_binary.columns]

          # Kết hợp với dataframe chính
          agg_df = agg_df.merge(problem_binary, on="concat('it001', username)", how='left')

          # Output the final dataframe
          agg_df.head()

    ```

**5. Bài tập - Nhận dạng chữ số viết tay**
- Mô tả về dữ liệu đóng góp: gồm 96 tấm chữ số viết tay, mỗi số có 9 tấm, riêng số 7 có 15 tấm.

- Giới thiệu bài toán: Bài toán này là bài toán phân loại số (digit classification) với input là 1 ảnh chứa 1 chữ số (0-9)  và output là một số (là số ở trong ảnh input).

- Dữ liệu mà nhóm dung để train mô hình là một bộ dữ liệu được download về trên GitHub của các nhóm đã public link trên GoogleClassroom, nhóm tiến hành crawl link này và tải về. Tạo thành dataset mà nhóm dùng để huấn luyện mô hình.

*Định nghĩa hàm trích xuất đặc trưng, hàm ngăn chặn các output dư thừa*

````python
# Hàm này ngăn chặn việc in ra màn hình trong quá trình xử lý, tránh gây tràn và lag máy khi in ra quá nhiều output không cần thiết
@contextmanager
def suppress_output():
    # Mở file devnull để sử dụng như là bộ đệm cho các output bị bỏ qua
    with open(os.devnull, 'w') as devnull:
        # Lưu lại các output ban đầu
        old_stdout = sys.stdout
        old_stderr = sys.stderr
        try:
            # Chuyển hướng sys.stdout và sys.stderr đến devnull để bỏ qua output
            sys.stdout = devnull
            sys.stderr = devnull
            # Thực hiện các xử lý trong khối with
            yield
        finally:
            # Khôi phục lại sys.stdout và sys.stderr về trạng thái ban đầu
            sys.stdout = old_stdout
            sys.stderr = old_stderr

# Hàm trích xuất đặc trưng
def extract_features(image_path, processor, model, device):
    try:
        # Mở và chuyển đổi ảnh thành định dạng RGB
        image = Image.open(image_path).convert('RGB')
        # Xử lý ảnh và chuyển đổi thành tensor, sau đó chuyển tensor này đến thiết bị được chỉ định (CPU hoặc GPU)
        inputs = processor(images=image, return_tensors="pt").to(device)
        # Tắt gradient tính toán để tiết kiệm bộ nhớ và tăng tốc độ xử lý
        with torch.no_grad():
            # Lấy đầu ra từ mô hình, đầu vào là các tensor đã xử lý
            outputs = model(**inputs)

        return outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
    except Exception as e:
        # In ra thông báo lỗi nếu có vấn đề xảy ra trong quá trình xử lý ảnh
        print(f"Error processing image {image_path}: {e}")
        return None  # Trả về None nếu có lỗi
````

````python
splits = range(1, 6)  # Split 1 đến 5
error_files = []

for split in splits:
    print(f"==============================Đang xử lý split {split}==============================")
    # Đọc dữ liệu train
    train_csv_path = f"{szWorkingDir}/MotocycleDataset-Splits-{split}-Train.csv"
    df_train = pd.read_csv(train_csv_path)
    df_train.columns = ['image_dir', 'category_id']


    # Sử dụng StratifiedShuffleSplit để tách 5% dữ liệu để chạy với notebook trên colab
    sss = StratifiedShuffleSplit(n_splits=1, test_size=0.95, random_state=42)
    for train_index, test_index in sss.split(df_train, df_train['category_id']):
        df_train, _ = df_train.iloc[train_index], df_train.iloc[test_index]

    features = []
    image_paths = []
    indices_to_keep = []
    split_error_files = []  # Danh sách file lỗi cho split hiện tại

    for index, path in tqdm(enumerate(df_train["image_dir"]), total=len(df_train["image_dir"])):
        full_path = os.path.join(szWorkingDir, path) # Tạo đường dẫn đầy đủ tới ảnh
        with suppress_output():
            feature = extract_features(full_path, processor, model, device)
        if feature is not None:
            features.append(feature) # Thêm đặc trưng vào danh sách features
            image_paths.append(full_path)
            indices_to_keep.append(index) # Thêm chỉ số của ảnh vào danh sách indices_to_keep
        else:
             # Nếu trích xuất đặc trưng thất bại, thêm đường dẫn đầy đủ của ảnh vào danh sách split_error_files
            split_error_files.append(full_path)

    features = np.array(features) # Chuyển danh sách features thành numpy array

    # Cập nhật df_train để giữ lại chỉ các hàng với hình ảnh đã được xử lý thành công
    df_train = df_train.iloc[indices_to_keep].reset_index(drop=True)

    print(f"Updated df_train with only successfully processed images for split {split}.")
    print(df_train.shape)
    print(features.shape)

    # Lưu lại features và df_train
    np.save(f"extracted_features-split{split}-train.npy", features)
    df_train.to_csv(f"S{split}_train.csv", index=False)

    # Đọc dữ liệu test
    test_csv_path = f"{szWorkingDir}/MotocycleDataset-Splits-{split}-Test.csv"
    df_test = pd.read_csv(test_csv_path)
    df_test.columns = ['image_dir', 'category_id']

    # Sử dụng StratifiedShuffleSplit để tách 5% dữ liệu để chạy với notebook trên colab
    for train_index, test_index in sss.split(df_test, df_test['category_id']):
        df_test, _ = df_test.iloc[train_index], df_test.iloc[test_index]

    features_test = []
    indices_to_keep_test = []

    for index, path in tqdm(enumerate(df_test["image_dir"]), total=len(df_test["image_dir"])):
        full_path = os.path.join(szWorkingDir, path)
        with suppress_output():
            feature = extract_features(full_path, processor, model, device)
        if feature is not None:
            features_test.append(feature) # Thêm đặc trưng vào danh sách features
            indices_to_keep_test.append(index) # Thêm chỉ số của ảnh vào danh sách indices_to_keep
        else:
             # Nếu trích xuất đặc trưng thất bại, thêm đường dẫn đầy đủ của ảnh vào danh sách split_error_files
            split_error_files.append(full_path)

    features_test = np.array(features_test) # Chuyển danh sách features thành numpy array
     # Cập nhật df_test để giữ lại chỉ các hàng với hình ảnh đã được xử lý thành công
    df_test = df_test.iloc[indices_to_keep_test].reset_index(drop=True)

    # Lưu lại features và df_test
    np.save(f"extracted_features-split{split}-test.npy", features_test)
    df_test.to_csv(f"S{split}_test.csv", index=False)

    print(f"Updated df_test with only successfully processed images for split {split}.")
    print(df_test.shape)
    print(features_test.shape)

    # In ra số lượng file lỗi trong split hiện tại
    print(f"Số lượng file lỗi trong split {split}: {len(split_error_files)}")

    # Thêm các file lỗi vào danh sách tổng thể
    error_files.extend(split_error_files)

    print(f"------------------------------Kết thúc xử lý split {split}------------------------------")

# Lưu danh sách các file bị lỗi vào file text
with open("error_files.txt", "w") as f:
    for file in error_files:
        f.write(file + "\n")

print("Danh sách các file bị lỗi đã được lưu vào 'error_files.txt'")
````

* *Định nghĩa class Voting Model để kết hợp 2 model LGBM và Catboost*


````python
data_path = '/content/'

# Các tham số cho mô hình LightGBM đã sử dụng optuna để tìm kiếm
params_lgb = {
    "boosting_type": "gbdt",
    "objective": "multiclass",
    "num_class": 5,
    "metric": "multi_logloss",
    "max_depth": 10,
    "learning_rate": 0.05,
    "n_estimators": 2000,
    "colsample_bytree": 0.8,
    "colsample_bynode": 0.8,
    "verbose": -1,
    "random_state": 5123,
    "reg_alpha": 0.1,
    "reg_lambda": 10,
    "extra_trees": True,
    'num_leaves': 64,
    "device": 'gpu',
    'subsample': 0.8,
    "extra_trees": True,
    "use_best_model": True
}

# Các tham số cho mô hình CatBoost đã sử dụng optuna để tìm kiếm
params_cat = {
    "iterations": 3807,
    "depth": 5,
    "learning_rate": 0.1265442460375865,
    "random_strength": 4.401688420764214e-05,
    "bagging_temperature": 0.014317301625633827,
    "border_count": 88,
    "l2_leaf_reg": 0.025648542214253733,
    "eval_metric": "MultiClass",
    "task_type": "GPU",
    "devices": "0",
    "random_seed": 1112,
    "use_best_model": True
}


# Định nghĩa class VotingModel để kết hợp predict của LGBM cà Catboost
class VotingModel(BaseEstimator, ClassifierMixin):
    def __init__(self, estimators):
        # Hàm khởi tạo, lưu trữ danh sách các mô hình (estimators) được truyền vào
        self.estimators = estimators

    def fit(self, X, y=None):
        # Hàm fit, hiện tại không thực hiện huấn luyện gì, chỉ trả về chính đối tượng
        return self

    def predict(self, X):
        # Tạo danh sách các dự đoán xác suất (predict_proba) từ mỗi estimator trong self.estimators
        y_preds_proba = [estimator.predict_proba(X) for estimator in self.estimators]
        # Tính trung bình của các dự đoán xác suất trên mỗi estimator
        avg_proba = np.mean(y_preds_proba, axis=0)
        # Trả về nhãn dự đoán cuối cùng bằng cách chọn nhãn có xác suất trung bình cao nhất
        return np.argmax(avg_proba, axis=1)
````

**Do chỉ chạy 5% data trên notebook BaoCaoDACK này nên các biểu đồ thể hiện không được tốt, em đã đính kèm các hình ảnh chạy với full data để có góc nhìn chính xác hơn**

````python
splits = range(1, 6)  # Split 1 đến 5
for split in splits:
    print(f"============================== Đang huấn luyện split {split} ==============================")

    # Tải features extraction đã lưu
    features_train = np.load(data_path + f"extracted_features-split{split}-train.npy")
    df_train = pd.read_csv(data_path + f"S{split}_train.csv")

    X_train = features_train
    y_train = df_train['category_id']

    # Áp dụng PCA để giảm số chiều và tạo thêm đặc trưng
    pca = PCA(n_components=260, random_state=42) # Tính toán tổng phương sai (thư viện sklearn hỗ trợ) -->  Tìm số lượng thành phần chính để giữ lại 98.5% phương sai --> xấp sỉ 260
    pca_features_train = pca.fit_transform(X_train)
    # Lưu PCA sử dụng joblib để evaluate sẽ sử dụng lại cho tập test
    joblib.dump(pca, f'pca_split_{split}.joblib')

    # Nối các đặc trưng PCA vào X_train
    X_train_pca = np.concatenate((pca_features_train, X_train), axis=1)

    # Sử dụng StratifiedShuffleSplit để lấy ra 10% làm validation data, đảm bảo rằng mỗi tập con của dữ liệu được tách ra có cùng tỷ lệ của các lớp như trong tập dữ liệu ban đầu
    sss = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=68)
    train_idx, eval_idx = next(sss.split(X_train_pca, y_train))

    X_train_split, X_eval = X_train_pca[train_idx], X_train_pca[eval_idx]
    y_train_split, y_eval = y_train.iloc[train_idx], y_train.iloc[eval_idx]

    # Huấn luyện mô hình LightGBM và lưu lại loss
    print("LightGBM:")
    model_lgb = LGBMClassifier(**params_lgb)
    model_lgb.fit(X_train_split, y_train_split, eval_set=[(X_eval, y_eval)], callbacks=[lgb.log_evaluation(500)])
    lgb_loss = model_lgb.evals_result_['valid_0']['multi_logloss']

    # Huấn luyện mô hình CatBoost và lưu lại loss
    print("CatBoost:")
    model_cat = CatBoostClassifier(**params_cat)
    model_cat.fit(X_train_split, y_train_split, eval_set=(X_eval, y_eval), verbose=500)
    cat_loss = model_cat.get_evals_result()['validation']['MultiClass']

    # Plot loss
    plt.figure(figsize=(12, 6))
    plt.plot(lgb_loss, label='LightGBM Loss')
    plt.plot(cat_loss, label='CatBoost Loss')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.title(f'Training Loss for Split {split}')
    plt.legend()
    plt.show()

    # Tạo mô hình voting
    voting_model = VotingModel(estimators=[model_lgb, model_cat])
    voting_model.fit(X_train_split, y_train_split)

    # Lưu mô hình voting cho mỗi split sử dụng joblib
    joblib.dump(voting_model, f'voting_model_split_{split}.joblib')
````

````python
acc = []
for split in splits:
    print(f"============================== Evaluating split {split} ==============================")

    # Tải dữ liệu đã lưu
    features_test = np.load(data_path + f"extracted_features-split{split}-test.npy")
    df_test = pd.read_csv(data_path + f"S{split}_test.csv")

    X_test = features_test
    y_test = df_test['category_id']

    # Tải PCA đã lưu và áp dụng trên features_test
    pca = joblib.load(f'pca_split_{split}.joblib')
    pca_features_test = pca.transform(X_test)

    # Nối các đặc trưng PCA vào X_test
    X_test_pca = np.concatenate((pca_features_test, X_test), axis=1)

    # Tải mô hình voting đã lưu
    voting_model = joblib.load(f'voting_model_split_{split}.joblib')

    # Dự đoán trên tập test
    y_pred = voting_model.predict(X_test_pca)

    # Tính độ chính xác
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy for Split {split}: {accuracy * 100:.4f}%')
    acc.append(accuracy)

    # Tính và hiển thị Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f'Confusion Matrix for Split {split}')
    plt.show()

average_accuracy = np.mean(acc)
print("Accuracy 5 split:", acc)
print(f"Trung bình Accuracy cho 5 split: {average_accuracy * 100:.4f}%")
````
